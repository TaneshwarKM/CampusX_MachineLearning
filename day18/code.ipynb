```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
soup.find_all('h1')[0].text
```

```python
for i in soup.find_all('h2'):
    print(i.text.strip())
```

```python
comp=soup.find_all('div',class_='companyCardWrapper__primaryInformation')
comp2=soup.find_all('div',class_='companyCardWrapper__tertiaryInformation')
comp2[0].find_all(class_='companyCardWrapper__ActionCount')[1].text.strip()
```

```python
name=[]
rating=[]
salary=[]
for i in range(len(comp)):
    name.append(comp[i].find_all('h2')[0].text.strip())
    rating.append(comp[i].find_all(class_='rating_star_container')[0].text.strip())
    salary.append(comp2[i].find_all(class_='companyCardWrapper__ActionCount')[1].text.strip())
```
