3. Onehotencoding using Sklearn

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:4], df.iloc[:,-1], test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(drop='first')
X_train_new=ohe.fit_transform(X_train[['fuel','owner']]).toarray()
X_test_new=ohe.transform(X_test[['fuel','owner']]).toarray()
# onehot encoding is done for the categorical variables fuel and owner. The drop='first' parameter is used to avoid the dummy variable trap.
```

```python
np.hstack([X_train[['brand','km_driven']].values, X_train_new]).shape

#p.hstack is used to horizontally stack the arrays. In this case, it combines the 'brand' and 'km_driven' columns from the training set with the one-hot encoded features for 'fuel' and 'owner'. The resulting shape of the combined array will reflect the total number of features after adding the one-hot encoded variables.
```

4.Onehotencoding with TopCategory

```python
counts=df['brand'].value_counts()
```

```python
df['brand'].nunique
threshold=100
```

```python
repl= counts[counts<=threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl,'uncommon')).sample(n=5)
#pd.get_dummies is used to perform one-hot encoding on the 'brand' column after replacing brands that occur less than or equal to the threshold (100) with the label 'uncommon'. This helps in reducing the number of unique categories in the 'brand' column, which can be beneficial for model performance and interpretability. The sample(n=5) method is used to display a random sample of 5 rows from the resulting DataFrame after one-hot encoding.
```
