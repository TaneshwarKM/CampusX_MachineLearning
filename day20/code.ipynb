```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
```

# 1. Categorical Data

a. Countplot

```python
sns.countplot(data=df, x='Embarked', hue='Survived')
#df['Survived'].value_counts().plot(kind='bar')
```

b. Piechart

```python
df['Sex'].value_counts().plot(kind='pie', autopct='%.2f')
```

# 2. Numerical Data

a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=50)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

b. Distplot

```python
sns.displot(data=df['Age'], kde=True, bins=30)
#We can this as Pdf= Probability Density Function. It is a function that describes the likelihood of a random variable to take on a particular value. The area under the curve of the PDF represents the probability of the random variable falling within a certain range of values.
```

c. Boxplot

```python
sns.boxplot(data=df, x='Age', y='Fare')
#boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”). It can also highlight outliers in the data.
```
